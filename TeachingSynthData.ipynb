## 📍 Module 1: The Biometric Fingerprint (The Body as Data)

We must first load the physical reality of the dancers into our environment. The dataset (`B1.csv`) contains rows of time-series data. Each row is a fraction of a second in the studio.

Before our AI can generate anything, it needs a blueprint. We call this the **Metadata**. Metadata is the schema that tells the machine learning model what data types (e.g., numerical brainwaves, categorical movement states) it is allowed to look at.

```python
import pandas as pd
from sdv.metadata import Metadata

# 1. Load the original biometric capture (The "Real" Data)
# Note: Ensure you have linked your Google Drive or uploaded the CSV locally.
try:
    real_brainwave_data = pd.read_csv('/content/drive/My Drive/brainwaves/B1.csv')
    print("🩰 Real biometric data loaded successfully. Here are the first 5 frames:")
    display(real_brainwave_data.head())
except Exception as e:
    print(f"Error loading data: {e}. Please check your file path.")

# 2. Detect the mathematical architecture of the dancers
metadata = Metadata.detect_from_dataframe(
    data=real_brainwave_data,
    table_name='dancer_biometrics'
)

# Visualize the blueprint
metadata.visualize()

```

## 👁️ Module 3: Evaluating the Unseen (The Visual Critique)

Now we must judge our creation. In synthetic data, we face a **Zero-Sum Game** between *Privacy* and *Utility*.

If our synthetic data scores **100%**, it means the AI memorized the dancers perfectly. This is called *Overfitting*. It is a catastrophic privacy failure because the data is just a clone of the real dancer. If it scores **20%**, it's abstract noise and useless for building our AR app.

We are aiming for the **Goldilocks Zone (75%–94%)**—an impressionistic representation of the dance. Let's run a diagnostic and create a visual "Glitch Art" comparison.

```python
from sdv.evaluation.single_table import evaluate_quality, get_column_pair_plot

# 1. Calculate the Statistical "Similarity Score"
quality_report = evaluate_quality(
    real_data=real_brainwave_data,
    synthetic_data=synthetic_data,
    metadata=metadata
)

# 2. Visual Critique: Data as Medium
# Let's plot the relationship between two specific brain regions during the dance.
columns_to_critique = ['AF3', 'F7']

fig = get_column_pair_plot(
    real_data=real_brainwave_data,
    synthetic_data=synthetic_data,
    column_names=columns_to_critique,
    metadata=metadata
)

fig.show()

```

## ⚖️ Module 4: Algorithmic Interventions (De-Biasing the Data)

**The Representation Trap:** Our AI was trained on exactly *two* classical ballet dancers. If a commercial developer uses our dataset to train a global fitness app, that app will likely fail to track users with larger bodies, physical disabilities, or different movement styles.

Video: [Dr. Joy Buolamwini reflects on decoding algorithmic bias and the future of AI](https://www.youtube.com/watch?v=6n3zvya2lHs$0)

We can use **Conditional Sampling** to fight this bias. We can force the generator to mathematically invent profiles that were rare or missing in the original studio recordings.

Let's force the AI to generate a dataset where half the dancers exhibit a very specific, rare high-focus brainwave state.

```python
from sdv.sampling import Condition

# 1. Define the intervention: We want a high concentration of intense brain activity
high_focus_condition = Condition(
    num_rows=250, # We are forcing exactly 250 rows to fit this profile
    column_values={'F7': (4500, float('inf'))} # F7 brainwave must be over 4500
)

# 2. Generate the intervened dataset
debiased_synthetic_data = synthesizer.sample_conditions(
    conditions=[high_focus_condition]
)

print("✊ Interventional dataset generated. We have manually expanded the representation:")
display(debiased_synthetic_data.head())

```

### 📚 Don't Panic: The "Syntax Translation" Glossary

As artists, you are already experts in structure, form, and movement. Computer science uses different words for these same concepts. Use this glossary to translate the Python code back into the language of the studio:

* **Dataset (`.csv`): The Body / The Raw Material.** This is the physical reality captured by the sensors. It contains the truth of the performance.
* **Model / Synthesizer: The Algorithmic Understudy.** The AI that we are teaching to dance. It starts with zero knowledge and must learn from the Dataset.
* **Training (`.fit()`): Rehearsals.** The process where the Understudy watches the data over and over to learn the statistical rhythm of the movements.
* **Overfitting (Similarity > 95%): Plagiarism / Surveillance.** The Understudy failed to learn the *concept* of the dance and instead just perfectly memorized the original dancer. This is a massive privacy failure.
* **Underfitting (Similarity < 50%): Abstract Noise.** The Understudy didn't rehearse enough. The output is chaotic, random, and ungrounded in physical reality.
* **Parameters: Director's Notes.** The settings you adjust (like `default_distribution`) to change the texture and style of how the AI learns.
* **Constraints & Conditioning: Choreography.** AI does not know physics or anatomy. When you write a `Condition` or a `ScalarRange`, you are actively choreographing what the machine is mathematically allowed to do.

### 🚨 "Don't Panic": Guide to Python Errors

When you are coding, you **will** get error messages. A giant block of red text does not mean you broke the computer, and it does not mean you are bad at data science. It is simply the Understudy telling you it doesn't understand your directorial notes.

Here is how to read an error message like a creative constraint:

**Rule #1: Always read the VERY LAST LINE first.**
Python prints the entire history of what went wrong, but the actual cause of the problem is always at the very bottom. Ignore the massive block of text and jump straight to the last sentence.

**Common SDV Errors in this Workshop:**

* **`FileNotFoundError` (or "No such file or directory")**
* *What it means:* The computer is blind. It cannot find your data.
* *How to fix it:* Did you mount your Google Drive? Did you misspell the name of the folder? Check your file path string (e.g., `'/content/drive/My Drive/brainwaves/B1.csv'`).


* **`KeyError: 'af3'`**
* *What it means:* You asked it to look at a column that doesn't exist. Python is case-sensitive!
* *How to fix it:* Check your capitalization. The dataset uses uppercase for brainwaves (`'AF3'`), so lowercase `'af3'` will crash the script.


* **`ValueError` during Conditional Sampling**
* *What it means:* You gave the model a mathematical paradox.
* *How to fix it:* Check your `Condition` logic. For example, did you ask it to generate an `AF3` value of `-500` when you previously set a `Constraint` that `AF3` must be above `0`?


* **`ModuleNotFoundError`**
* *What it means:* You forgot to install your tools.
* *How to fix it:* Scroll all the way back up to the setup block and make sure you ran the `!pip install sdv` cell.
